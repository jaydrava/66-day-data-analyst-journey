```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

pd.set_option('display.max_columns', None)
plt.style.use('default')
```

```python
customers = pd.read_csv('../data/customers.csv')
products = pd.read_csv('../data/products.csv')
orders = pd.read_csv('../data/orders.csv')
order_items = pd.read_csv('../data/order_items.csv')
```

```python
# check duplicate customers by customer_id
duplicate_customers = customers[customers.duplicated(subset='customer_id')]
duplicate_customers.shape
```

```python
# Check completed orders with zero or negative value
invalid_orders = orders[
    (orders['order_status'] == 'Completed') &
    (orders['total_amount'] <= 0)
]
invalid_orders.shape
invalid_orders.head()
```

```python
customers.isna().sum() 
```

```python
sales_by_category = (
    pd.merge(order_items, products, on='product_id', how='inner')
    .groupby('category', as_index=False)['subtotal']
    .sum()
    .sort_values('subtotal', ascending=False)
)
print(sales_by_category)
```

```python

plt.bar(sales_by_category['category'], sales_by_category['subtotal'])

plt.xlabel('Product Category')
plt.ylabel('Total Revenue')
plt.title('Revenue by Product Category')

plt.show()
```

```python
order_with_status = pd.merge(order_items, orders, on='order_id', how='inner')
merge_df = pd.merge(order_with_status, products, on='product_id', how='inner')
filter_df = merge_df[~((merge_df['order_status'] == 'completed') & (merge_df['subtotal'] == 0))]
result = (
    filter_df
    .groupby(['category', 'product_name'],as_index=False)['subtotal'].sum()
    .rename(columns={'subtotal':'total_revenue'})
    .sort_values('total_revenue', ascending=False)
)
print(result)
```

```python
result['Rank']=result['total_revenue'].rank(method='dense', ascending=False).astype(int)
result= result.sort_values(by='Rank')
result['cumulative_revenue'] = result['total_revenue'].cumsum()
total_revenue_all_products = result['total_revenue'].sum()
result['cumulative_revenue_pct']= (result['cumulative_revenue']/total_revenue_all_products)*100
print(result)
```

```python
plt.figure(figsize=(10,6))
plt.plot(result['Rank'],result['cumulative_revenue_pct'],
        marker='o', 
         color='green', 
         linewidth=2, 
         markersize=8,
         label='Revenue')
plt.xlabel('Rank',fontsize=12)
plt.ylabel('cumulative_revenue_pct',fontsize=12)
plt.title('Cumulative Revenue by Product',fontsize=14)
plt.show()
```

```python
customers_by_sales = pd.merge(orders,customers,on='customer_id', how='inner')
customer_result = (customers_by_sales.groupby(['customer_name','customer_segment'],as_index=False)['total_amount'].sum()
                   .sort_values('total_amount',ascending=False)
                   .head(10)
                  )
print(customer_result)
```

```python
plt.barh(customer_result['customer_name'],customer_result['total_amount'])
plt.xlabel('customer name',fontsize=12)
plt.ylabel('total revenue',fontsize=12)
plt.title('Customers by Revenue',fontsize=14)
plt.show()
```
