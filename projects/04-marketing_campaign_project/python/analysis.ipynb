```python
import pandas as pd

df = pd.read_csv("../data/daily_revenue_per_email.csv")
df.head()
```

```python
df.shape
df["campaign_type"].value_counts()
df.isna().sum()
print(df.shape)
print(df["campaign_type"].value_counts())
```

```python
campaign_a = df[df['campaign_type'] == 'A']['daily_revenue_per_email'].to_numpy()
campaign_b = df[df['campaign_type'] == 'B']['daily_revenue_per_email'].to_numpy()
print(f"Campaign A lenght is : {len(campaign_a)} rows")
print(f"Campaign B lenght is : {len(campaign_b)} rows")
print(f"Campaign A Average is : {campaign_a.mean().round(4)}")
print(f"Campaign B Average is : {campaign_b.mean().round(4)}")
```

```python
from scipy import stats
t_statistic, p_value = stats.ttest_ind(campaign_a,campaign_b,equal_var=False)
print(f"\nT-test results:")
print(f"T-statistic: {t_statistic:.4f}")
print(f"P-value: {p_value:.4f}")
if p_value < 0.05:
    print("\n✅ Campaign B is the winner for revenue per email, and the difference is statistically significant (p < 0.05)")
    print(f"   P-value ({p_value:.4f}) < 0.05")
    print(f"   Improvement is statistically significant")
    print(f"   Recommend: Shift more volume to Campaign B as the default, and keep A targeted for high-converting segments.")
else:
    print("\n❌ NO SIGNIFICANT EFFECT")
    print(f"   P-value ({p_value:.4f}) > 0.05")
    print(f"   Improvement might be random")
    print(f"   Recommend: Don't invest in campaign")
    
```
